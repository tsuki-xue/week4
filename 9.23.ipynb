```python
import matplotlib.pyplot as plt
import numpy as np
rng=np.random.RandomState(10)
```

```python
k=rng.rand(100)
```

```python
plt.scatter(k,f,c=rng.rand(100),sizes=1000*rng.rand(100),alpha=0.3,cmap='viridis')
```

```python
categ=['beer','wine','spring','alcoh']
va=[10,20,50,35]
```

```python
plt.bar(categ,va,color='pink',width=0.5)
plt.xlabel('jiu')
plt.ylabel('sail')
plt.title('jiu sail')
```

```python
x=np.random.normal(0,0.8,1000)
y=np.random.normal(-3,1,1000)
kwargs=dict(alpha=0.5,bins=20,density=True)
```

```python
plt.hist(x,**kwargs)
plt.hist(y,**kwargs)
```

```python
x = [1,2,3,4,5,6]
y = ['dog','cat','elephant','mouse','human','snake']
plt.pie(x,labels=y, autopct='%1.1f%%')
plt.title('pie')
plt.xlabel('x-label')
plt.ylabel('y-label')
```
